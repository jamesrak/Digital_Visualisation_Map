```python
import pandas as pd
```

```python
df = pd.read_csv('website/data/ghg_emission_2020.csv')
df
```

```python
df.columns
```

```python
top10_df = df.sort_values(by=['CO2_Grand Total'],ascending=False).head(10)
top10_df = top10_df.round(0)
top10_df = top10_df.rename({
    'CO2_Agriculture':'Agriculture',
    'CO2_Commercial':'Commercial',
    'CO2_Domestic' : 'Domestic',
    'CO2_Industry' : 'Industry',
    'CO2_Public Sector' : 'Public Sector',
    'CO2_Transport': 'Transport'
},axis=1)
top10_df
```

```python
top10_df.to_csv('ghg_website/data/top10_co2_emission.csv',index=False)
```
