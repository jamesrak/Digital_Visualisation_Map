# GHG Emission Data
Aggregate data by sector

```python
df = pd.read_csv('dataset/uk_ghg_emission/uk-local-authority-ghg-emissions-2020-dataset.csv')
df.info()
```

```python
df.head()
```

```python
#Exclude LULUCF and Waste Management Sectors due to existing of many unknown data
selected_df = df[(df['Calendar Year']>=2016) & (df['Greenhouse gas'] == 'CO2') & (df['LA GHG Sector'] != 'LULUCF') & (df['LA GHG Sector'] != 'Waste management')]
selected_df.head()
```

```python
agg_df = selected_df.groupby(['Local Authority Code','Region','Second Tier Authority','Local Authority','Calendar Year','LA GHG Sector','Greenhouse gas'])[['Territorial emissions (kt CO2e)']].sum().reset_index()
agg_df
```

```python
agg_df.columns
```

```python
total_df = agg_df.groupby(['Local Authority Code','Region','Second Tier Authority','Local Authority','Calendar Year','Greenhouse gas'])[['Territorial emissions (kt CO2e)']].sum().reset_index()
total_df['LA GHG Sector'] = 'Grand Total'
total_df = total_df[agg_df.columns]
total_df
```

```python
new_df = pd.concat([agg_df,total_df])
new_df = new_df.rename({'LA GHG Sector':'Sector'},axis=1)
new_df
```

```python
new_df.columns
```

```python
## Filter only 2020 data and pivot table
pivot_df = new_df[(new_df['Calendar Year'] == 2020)].pivot(index=['Local Authority Code', 'Region', 'Second Tier Authority',
       'Local Authority', 'Calendar Year','Greenhouse gas'],columns=['Sector'],values=['Territorial emissions (kt CO2e)']).reset_index()
pivot_df.columns = ['Local Authority Code', 'Region', 'Second Tier Authority',
       'Local Authority', 'Calendar Year','Greenhouse gas',"Agriculture","Commercial","Domestic","Grand Total","Industry","Public Sector","Transport"]
pivot_df = pivot_df[['Local Authority Code', 'Region', 'Second Tier Authority',
       'Local Authority', 'Calendar Year','Greenhouse gas',"Agriculture","Commercial","Domestic","Industry","Public Sector","Transport","Grand Total"]]
pivot_df = pivot_df.round(0)
pivot_df
```

```python
#extract sector that produce CO2 most in each local area
pivot_df['TopSector'] = pivot_df[["Agriculture","Commercial","Domestic","Industry","Public Sector","Transport"]].idxmax(axis=1)
pivot_df
```

```python
pivot_df.groupby(['TopSector']).count()[['Region']]
```

```python
pivot_df.info()
```

```python
pivot_df.to_csv('ghg_website/data/ghg_emission_v2_2020.csv',index=False)
```

# UK Local Authorities Boundaries

```python
import geopandas as gpd
boundary_df = gpd.read_file('dataset/uk_ghg_emission/uk-local-authority-district.geojson')
boundary_df.info()
```

```python
boundary_df.head()
```

```python
join_df = pivot_df.set_index('Local Authority Code').join(boundary_df[['reference','geometry']].set_index('reference'),how='left').reset_index()
join_df.info()
```

```python
join_df = join_df[pd.notnull(join_df['geometry'])]
join_df = join_df.rename({'index':'Local Authority Code'},axis=1)
join_df.info()
```
